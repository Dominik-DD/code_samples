```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("scrape.csv", encoding = 'utf-16')
```

```python
df
```

```python
#replacing NaN (and other types) with 0´s
```

```python
df['Sales'] = df['Sales'].str.replace('\xa0','')
```

```python
df['Sales'] = df['Sales'].fillna(0)
```

```python
df['Sales'] = df['Sales'].astype(int)
```

```python
df['Employees'] = df['Employees'].str.replace('nezistený','0')
```

```python
df['Employees'] = df['Employees'].str.split("-") 
```

```python
for i in range(len(df['Employees'])):
    if len(df['Employees'][i]) > 1: # if len > 1 ... e.g. [3,4] after spilitng from "3-4"
        for y in range(len(df['Employees'][i])): 
            df['Employees'][i][y] = int(df['Employees'][i][y]) # conversion to int
        df['Employees'][i] = (df['Employees'][i][0]+df['Employees'][i][-1])//2 # average with floor division
            
    else:
        if df['Employees'][i][-1] == '+': 
            df['Employees'][i] = df['Employees'][i][:-1]
        else:    
            df['Employees'][i] = "".join(df['Employees'][i])
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.reset_index(inplace = True, drop = True)
```
